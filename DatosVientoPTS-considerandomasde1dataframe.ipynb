```python
def subdata(path_name):
    # Nombre del archivo de datos
    #nombrearchivo = "C:/Users/enclab01/Desktop/THESIS/Datos Viento/WwWind/pts/2017/006/pts17006.18w"

    titles=['HT',      'SPD',      'DIR',  'MET_QC'   , 'RAD1',      'RAD2',      'RAD3',      'CNT1',      'CNT2',      'CNT3',      'SNR1',      'SNR2',      'SNR3',  'QC1',      'QC2',      'QC3'  ]
    data=pd.read_table(path_name) 


    'Here we count how many $ symbols we have in the data in order to correctly identify the datasets per hour of the events'
    counter=0
    limits=[]
    for i in np.arange(0,len(data)):
        if data.iloc[np.arange(i,i+1)][' PTS'][i]=='$':
            counter=counter+1
            limits.append(i)

    if len(limits)>2:
        print('TENEMOS UN DIA CON MAS DE 2 DATOS POR HORA')
        
    
    list_of_subdata=[]
        
    for i in np.arange(0,counter):
        if i==0:
        
            subdata=pd.read_table(nombrearchivo, skiprows=11 ,nrows=limits[i]-9,header=None,names=titles, sep='\s+') #,parse_dates=['t_utc'])
            list_of_subdata.append(subdata)
            #print(subdata)
        else:
        
            subdata=pd.read_table(nombrearchivo, skiprows=limits[i-1]+13,nrows=limits[i]-limits[i-1]-11 ,header=None,names=titles, sep='\s+') #,parse_dates=['t_utc'])
            list_of_subdata.append(subdata)
            #print(subdata)    
    
    if len(list_of_subdata)==1:
        return list_of_subdata[0]
    elif len(list_of_subdata)==2:
        return list_of_subdata[0], list_of_subdata[1]
    elif len(list_of_subdata)==3:
        return list_of_subdata[0], list_of_subdata[1], list_of_subdata[2]
    else:
        return ('WE HAVE AN ERROR')
    
    
```
